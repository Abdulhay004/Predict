# **main absolute error**

image.png

```python
# import libraries
from sklearn.metrics import mean_absolute_error as MAE
import numpy as np

num1 = 10
num2 = 20

a = np.arange(num1)
b = np.arange(1, num2, 2)

print("result of main absolute error: [{}]".format(MAE(a, b)).upper())
```

```python
my_maE = 1/len(a)*sum((abs(a - b)))
print("my MAE: [{}]".format(my_maE))
```

# **r2_score**

image.png

```python
from sklearn.metrics import r2_score as r2

print("result of r2_score: [{}]".format(r2(a,b)).upper())
```

```python
x = sum((b-a)**2)
y = sum((a-np.mean(a))**2)

print("my r2: [{}]".format(1-x/y).upper())
```
